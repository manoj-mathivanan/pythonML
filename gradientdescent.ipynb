```
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x1, x2):
    return 8 + (x1**2) / 2 + 2 / (x1 * x2) + 6 * x2

# Define the gradient of the function
def grad_f(x1, x2):
    grad_x1 = x1 - 2 / (x1**2 * x2)
    grad_x2 = -2 / (x1 * x2**2) + 6
    return np.array([grad_x1, grad_x2])

# Gradient descent algorithm that tracks the function values
def gradient_descent(learning_rate, initial_guess, epsilon=1e-6, max_iters=10000):
    x = np.array(initial_guess)
    f_values = []  # Track function values at each iteration

    for i in range(max_iters):
        f_values.append(f(x[0], x[1]))  # Store function value
        grad = grad_f(x[0], x[1])
        new_x = x - learning_rate * grad

        # Stopping criterion: small change in function value
        if np.linalg.norm(f(new_x[0], new_x[1]) - f(x[0], x[1])) < epsilon:
            break

        x = new_x

    f_values.append(f(x[0], x[1]))  # Append the last function value
    return x, f(x[0], x[1]), i, f_values  # Return final x values, function value, iterations, and function values

# Parameters
initial_guess = [1.0, 1.0]  # Starting point
learning_rates = [0.001, 0.005, 0.01, 0.05]  # List of learning rates to experiment with

# Plotting the function values over iterations
plt.figure(figsize=(10, 6))

for lr in learning_rates:
    final_x, final_f, iterations, f_values = gradient_descent(learning_rate=lr, initial_guess=initial_guess)
    plt.plot(f_values, label=f'Learning rate = {lr}')

# Graph details
plt.title('Function Value over Iterations for Different Learning Rates')
plt.xlabel('Iterations')
plt.ylabel('Function Value')
plt.legend()
plt.grid(True)
plt.show()
```
